# Extra: Copy or reference?

* basic types (int, float, string) **cannot** be modified without 'relocating', so in practice they behave similarly to how they work in, say, C++ (disregarding references/pointers)
* we can check if two labels refer to the same memory by using the **is** operator (ant not ==)
* we can also get the unique id (often returning something pertaining to the memory address) of a variable
* other types we'll see later will behave differently, so it leads to easy to miss bugs!



```python
b = 50
a = b

print(id(a), id(b))

print(a is b) # True, labels a and b refer to the same memory location
b += 10 # b = b + 10
```

```python
print(a is b) # False
print(id(a), id(b))
print(a, b) # different values, only b gets changed
```
